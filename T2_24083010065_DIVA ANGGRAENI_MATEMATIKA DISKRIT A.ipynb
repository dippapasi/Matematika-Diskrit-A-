**Tentukan prioritas/nilai beban untuk A, B, C, D sebagai matrix P; pilih empat orang (baris) dengan nilai terbesar dari matrix C.P**

```python
C = df.iloc[:, 2:].to_numpy()
P = np.array([1.0, 0.8, 0.6, 0.9])

skor = C @ P
df['Skor'] = skor
df_sorted = df.sort_values(by='Skor', ascending=False)

# Top 4 mahasiswa dengan skor terbesar
top4 = df_sorted.head(4)

print("Empat mahasiswa dengan skor tertinggi:")
print(top4[['NAMA', 'Skor']])
```

Nilai beban (P) yang saya gunakan untuk setiap kriteria yaitu adil, amanah (1.0); suka membantu (0.8); nyambung kalau ngobrol (0.6); dan asyik, menarik (0.9). Empat mahasiswa dengan skor tertinggi ditentukan dari hasil penjumlahan matriks C.P dari setiap kriteria siswa. Dengan begitu berdasarkan total skor tersebut, diambil empat mahasiswa yang memiliki skor tertinggi untuk ditampilkan.

**Hitung matriks kovarian dan korelasi**
* cov_mat = np.cov(kelasku, rowvar=False, ddof=1)
* corr_mat = np.corrcoef(kelasku, rowvar=False)

```python
kelasku = df.iloc[:, 2:]
cov_df = pd.DataFrame(np.cov(kelasku, rowvar=False, ddof=1), index=kelasku.columns, columns=kelasku.columns)
corr_df = pd.DataFrame(np.corrcoef(kelasku, rowvar=False), index=kelasku.columns, columns=kelasku.columns)

print("Matriks Kovarian")
display(cov_df)
print("Matriks Korelasi")
display(corr_df)
```

**Matriks kovarian:** hanya menunjukkan arah hubungan antar variabel (positif/negatif) sehingga tidak dapat diukur langsung seberapa kuat hubungan sebab  dipengaruhi skala data. Semua kriteria memiliki hubungan positif dengan skor yang berarti semakin tinggi nilai kriteria maka semakin tinggi juga skornya.

**Matriks korelasi:** memiliki nilai antara -1 sampai 1, sehingga lebih mudah diketahui seberapa kuat hubungan antar variabel. Dari hasil tersebut terlihat bahwa kriteria adil, amanah memiliki korelasi paling kuat dengan skor (0.65) yang berarti kriteria tersebut paling memengaruhi skor.

**Hitung cosine similarity antar siswa**
* norms = np.linalg.norm(X_matrix, axis=1, keepdims=True)
* cosine_sim = (X_matrix @ X_matrix.T) / (norms @ norms.T)
* cosine_df = pd.DataFrame(cosine_sim, index=students, columns=students)


```python
matrix = df.iloc[:, 2:6].to_numpy()

norms = np.linalg.norm(matrix, axis=1, keepdims=True)
cosine_sim = (matrix @ matrix.T) / (norms @ norms.T)
cosine_df = pd.DataFrame(cosine_sim, index=df['NAMA'], columns=df['NAMA'])

print("Cosine Similarity Antar Siswa")
cosine_df
```

Pada cosine similarity ini mengukur kemiripan antar siswa berdasarkan nilai kriterianya. Contohnya pada baris pertama, Madina dengan dirinya sendiri nilainya 1.000000 pasti sama persis. Jika Madina dibandingkan Najwa Sofia, tingkat kemiripan kriterianya 0.906240 masih mirip namun masih lebih rendah dibandingkan Madina dengan Maria yang nilainya 0.993781 (hampir mirip).

**Analisis tambahan (clustering)**

```python
from sklearn.cluster import KMeans
import numpy as np

X = df[["Skor"]].to_numpy()
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(X)
print("Cluster mahasiswa berdasarkan skor:")
print(df[["NAMA", "Skor", "Cluster"]].sort_values("Skor", ascending=False))

# Keterangan rata-rata skor tiap cluster
print("\nRata-rata skor per cluster:")
print(df.groupby("Cluster")["Skor"].mean())
```

Clustering ini mengelompokkan mahasiswa berdasarkan skor C.P yang sudah dihitung sebelumnya. Cluster 0 memiliki rata-rata skor 2.2860, lebih rendah dibandingkan cluster 1 yang memiliki rata-rata skor 2.6625. Mahasiswa dengan cluster 1 dianggap sebagai ketua kelompok sebab berdasarkan skor penilaian kriteria ia lebih unggul dibandingkan yang lain, sehingga layak menjadi pemimpin. Sedangkan, mahasiswa dengan cluster 0 dianggap sebagai anggota kelompok sebab skor penilaian kriteria lebih rendah dibandingkan dengan cluster 1. Dengan begitu, clustering ini membantu dalam pengelompokan mahasiswa secara objektif berdasarkan skor penilaian karakter yang diperoleh.
